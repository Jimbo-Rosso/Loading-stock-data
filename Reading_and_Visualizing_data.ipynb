We will look at stock prices from March 2013 to April 2014

```python
start_date = dt.date(2013,3,3)  #year, month, day
end_date = dt.date(2014,4,3) 
print(start_date.strftime("%Y"))
print(end_date)
```

## <font color=blue>Reading data 

 Let's get Citi and Chase (JPM) stock data; their's ticker symbols are C and JPM correspondingly.   
 To read the data, the first argument is the series we want, second is the source ("yahoo" or "google" Finance ), third is the start date, fourth is the end date


```python
Citi = web.DataReader("C", "yahoo", start_date, end_date)
```

```python
Chase = web.DataReader("JPM", "yahoo", start_date, end_date)
```

## <font color=blue>We visualize the "Close" column

As you can see the activity looks similar, with Citi's activity below JPM group

```python

plt.title('Stock-Close')
plt.grid(True)
Citi["Close"].plot(grid = True, color="red",  linewidth=2.5, linestyle="-", label="Citi")
Chase["Close"].plot(grid = True,color="blue",  linewidth=2.5, linestyle="-", label="Chase")
plt.legend(loc='upper left')
```
